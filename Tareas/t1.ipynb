Demuestre lo siguiente

1. Si $f_1$ es $O(g_1)$ y $f_2$ es $O(g_2)$ entonces $(f_1+f_2)$ es $O(max{|g_1|,|g_2|})$
2. Si $f_1$ es $O(g_1)$ y $f_2$ es $O(g_2)$ entonces $(f_1f_2)$ es $O(g_1g_2)$
3. ¿Cuál es la complejidad del siguiente algoritmo(O)?

for (i = 0; i < N; i++) {

    for (j = 0; j < M; j++) {
        print(i,j)
    }

}


---

1.$$
f_{1}+f_{2}=O\left(g_{1}(n)\right)+O\left(g_{2}(n)\right)
$$

$$
O\left(g_{1}(n)\right)+O\left(g_{2}(n)\right)=c_{1} g_{1}(n)+c_{2} g_{2}(n)
$$

Si se cumple  $
g_{1} \leq g_{2}
$


$$
c_{1} g_{1}(n)+c_{2} g_{2}(n) \leq c_{1} g_{2}(n)+c_{2} g_{2}(n)
$$

$$
c_{1} g_{1}(n)+c_{2} g_{2}(n) \leq\left(c_{1}+c_{2}\right) g_{2}(n)
$$

Sea $
c_{1}+c_{2}=c
$


$$
c_{1} g_{1}(n)+c_{2} g_{2}(n) \leq c g_{2}(n) \leq \operatorname{cmax}\left(g_{1}, g_{2}\right)
$$

$$
0 \leq c, g_{1}(n)+c_{2} g_{2}(n) \leq \operatorname{cmax}\left(g_{1}, g_{2}\right)
$$

Aplicando la defición de $O(g(n))$

$$
\therefore f_{1}+f_{2}=O\left(\max \left(g_{1}, g_{2}\right)\right)
$$

2.
$$
f_{1} f_{2}=O\left(g_{1}\right) O\left(g_{2}\right)=\left(c_{1} g_{1}(n)\right)\left(c_{2} g_{2}(n)\right)=c_{1} c_{2} g_{1}(n) g_{2}(n)
$$

Sea $
c_{1}c_{2}=c
$

$$
f_{1} f_{2}= c g_{1}(n) g_{2}(n)=O\left(g_1 g_{2}\right)
$$


$$
\therefore f_{1}f_{2}=O\left(g_1 g_{2}\right)
$$

```python
import numpy as np
from time import time
import pylab as pb
```

3. Un loop es un método cuyo tiempo de ejecución aumenta linealmente con n. Por ejemplo

```python
def ejemplo(N):
    for i in range(N):
        i+1      
```

```python
def On(funct,lista,N):
    for i in range(N):
        t_ini=time()*1000.0 #time() devuelve el tiempo en segundos desde el epoch multplicamos por 1000 
        #para obtener el tiempo en milisegundos 
        funct(i)
        t_fin=time()*1000.0
        t_total=t_fin-t_ini
        lista.append(t_total)
    return t_total
```

```python
salida=[]
On(ejemplo,salida,100)
```

```python
r=[i for i in range(len(salida))] # lista de valores para el eje X
pb.title('Complejidad en tiempo')
pb.xlabel('n')
pb.ylabel('Tiempo de ejecución (ms)')
pb.plot(r,salida,label='función')

pb.grid()
pb.legend()
pb.show()
```

Por lo tanto la complejidad de un ciclo for es $O(n)$.

En nuestro caso tenemos dos ciclos anidados, de modo que para cada valor N del primer ciclo, el segundo ciclo se ejecuta M veces. Por lo tanto, la complejidad total estará dada por el producto de las complejidades de ambos ciclos, es decir $O(n)O(m)=O\left(g_1 g_{2}\right)$

```python
def f(N,M):
    for i in range(N): 

        for j in range(M):
            print(i,j)
```

```python
def On(funct,lista,N):
    for i in range(N):
        t_ini=time()*1000.0 #time() devuelve el tiempo en segundos desde el epoch multplicamos por 1000 
        #para obtener el tiempo en milisegundos 
        funct(i,i)
        t_fin=time()*1000.0
        t_total=t_fin-t_ini
        lista.append(t_total)
    return t_total
```

```python
tiempos=[]
N=11
On(f,tiempos,N)
```
